## 1-4. Figure과 Axes

matplotlib를 시작하고 가장 어려운 부분 중 하나는 Figure와 Axes입니다.
우선 기본적인 문법으로 틀만 출력해보겠습니다.

```python
import numpy as np
import matplotlib as mpl 
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, 1)
plt.show()
```

`plt.subplots()`라는 함수를 사용해 우리는 2가지 객체를 받을 수 있습니다. 
(`plt.subplots()` 대신 `plt.subplot()` 함수를 사용하면 Axes만 받을 수도 있습니다.)

1. **Figure** : plot의 최상위 레이어
2. **Axes** : Figure의 일부분

이렇게만 보면 왜 굳이 나누었는지 궁금합니다. 

좀 더 디테일하게 살펴봅시다.

- **Figure**는 **subplots, subaxis, subtitiles, legends** 등 plot의 모든 것을 다룹니다.
- **Axes**는 **x-axis limit, y-axis limit, label, graph 종류** 등 subplot의 모든 것을 다룹니다.

Figure는 나타나는 모든 것을 다루는 반면, Axes는 서브플롯 안에 나타나는 모든 것을 다룬다고 생각하면 됩니다. 

```python
ax = plt.subplot()
```

그래서 똑같은 사각형 2개를 그리더라도 다음과 같이 2가지 방법이 있습니다.

```python
# subplots 사용 (1)

fig, axes = plt.subplots(1, 2) # 1 by 2
plt.show()
```

Python 문법을 조금 활용하여 axes 반환을 개별적으로 받을 수도 있습비다.

```python
# subplots 사용 (2)

fig, (ax1, ax2) = plt.subplots(1, 2) # 1 by 2
plt.show()
```

subplot을 사용하면 1개의 Ax를 반환하니 다음과 같이 할 수 있습니다.

```python
# subplot으로 개별적 그리기

ax1 = plt.subplot(1, 2, 1) # 1 by 2의 1번째
ax2 = plt.subplot(1, 2, 2) # 1 by 2의 2번째
plt.show()
```

`plt.subplots`은 한번에 동일한 크기의 subplot을 한 번에 그리기 쉽습니다.
`plt.subplot`은 개별적으로 그려 서로 다른 크기의 subplot을 그릴 수 있다는 장점이 있습니다. 

각각의 장점을 이용한 그래프를 그려보겠습니다.

```python
# subplots로 12개 그리기

fig, axes = plt.subplots(3, 4)
axes[1][2].plot([1, 2, 3, 4]) # subplots으로 반환한 axes는 배열처럼 접근가능
axes[2][1].bar([1, 2, 3, 4], [1, 2, 3, 4])
plt.tight_layout() # subplot간 margin(여백으로 이해하시면 됩니다.)이 겹치지 않게 하는 함수
plt.show()
```

```python
# subplot으로 다른 크기 그리기

ax221 = plt.subplot(2, 2, 1)
ax222 = plt.subplot(2, 2, 2)
ax212 = plt.subplot(2, 1, 2)

ax221.scatter([1, 2, 3], [3, 2, 1])
ax212.plot([1,2,3,4,5]) # subplot으로 만든 ax는 각각을 개별적으로 사용할 수 있음

plt.tight_layout()
plt.show()
```

물론 subplot을 여러개 그리는 더 고급진 방법이 있으나 이는 후에 소개하도록 하겠습니다.

그리고 ax를 받으면 이제 이 객체에는 여러 개의 그래프를 함께 그릴 수도 있습니다.

```python
ax = plt.subplot()

x = np.array([1, 2, 3, 4, 5])
y = x**2

ax.bar(x, y)
ax.plot(x, y)
plt.show()
```

이렇게 한 plot에는 다양한 종류의 graph를 그려 내용을 추가할 수 있습니다. 

하지만 지금처럼 아에 색이 같으면 그래프가 서로 중첩되며 가독성이 떨어집니다. 
이런 상황을 방지하기 위해 색, 투명도 등을 조절할 수 있습니다.
